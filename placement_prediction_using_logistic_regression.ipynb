```python
from sklearn.linear_model import LogisticRegression

# Assuming X_train and y_train are your training data
model = LogisticRegression()
model.fit(X_train, y_train)

# Get all parameters using vars() and filter out private attributes (starting with '_')
all_params = {key: value for key, value in vars(model).items() if not key.startswith('_')}

# Print all parameters
for param_name, param_value in all_params.items():
    print(f"{param_name}: {param_value}")
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
h = 0.02
x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))


Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
y_train_values = y_train.values.ravel()  # Assuming y_train is a DataFrame or Series


plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train_values, edgecolors='k')
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.title('Logistic Regression')

plt.show()
```
